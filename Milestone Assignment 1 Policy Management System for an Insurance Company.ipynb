```python
#Class for Policyholder
class Policyholder:
    def __init__(self, name, email):
        self.name = name
        self.email = email
        self.status = "active"
        self.products = []

#Class for Products
class Product:
    def __init__(self, name, premium):
        self.name = name
        self.premium = premium
        self.status = "active"

#Class for Payment
class Payment:
    def __init__(self, policyholder, product, amount):
        self.policyholder = policyholder
        self.product = product
        self.amount = amount
        #self.date = datetime.now()

# Define a PolicyManager class to manage policymanagers, products, and payments
class PolicyManager:
    def __init__(self):
        self.policyholders = []
        self.products = []

    def add_policyholder(self, name, email):
        policyholder = Policyholder(name, email)
        self.policyholders.append(policyholder)

    def suspend_policyholder(self, name):
        for policyholder in self.policyholders:
            if policyholder.name == name:
                policyholder.status = "suspended"

    def add_product(self, name, premium):
        product = Product(name, premium)
        self.products.append(product)

    def suspend_product(self, name):
        for product in self.products:
            if product.name == name:
                product.status = "suspended"

    def register_member(self, policyholder_name, product_name):
        for policyholder in self.policyholders:
            if policyholder.name == policyholder_name:
                for product in self.products:
                    if product.name == product_name:
                        policyholder.products.append(product)

    def process_payment(self, policyholder_name, product_name, amount):
        for policyholder in self.policyholders:
            if policyholder.name == policyholder_name:
                for product in policyholder.products:
                    if product.name == product_name:
                        payment = Payment(policyholder, product, amount)
                        # Process payment logic here
                        print(f" {policyholder.name} has made payment for {product.name}")

    def display_account_details(self, policyholder_name):
        for policyholder in self.policyholders:
            if policyholder.name == policyholder_name:
                print(f"Policyholder: {policyholder.name}")
                print(f"Email: {policyholder.email}")
                print(f"Status: {policyholder.status}")
                print("Products:")
                for product in policyholder.products:
                    print(f"  - {product.name} ({product.premium})")
                print()

# Create a PolicyManager instance
policy_manager = PolicyManager()

# Add policyholders
policy_manager.add_policyholder("Dakolo Junior", "Junosdakolo@gmail.com")
policy_manager.add_policyholder("Dakolo Emmanuel", "Junosdakolo@yahoo.com")

# Add products
policy_manager.add_product("Life Insurance", 100)
policy_manager.add_product("Health Insurance", 50)

# Register members
policy_manager.register_member("Dakolo Junior", "Life Insurance")
policy_manager.register_member("Dakolo Emmanuel", "Health Insurance")

# Process payments
policy_manager.process_payment("Dakolo Junior", "Life Insurance", 100)
policy_manager.process_payment("Dakolo Emmanuel", "Health Insurance", 50)

# Display account details
policy_manager.display_account_details("Dakolo Junior")
policy_manager.display_account_details("Dakolo Emmanuel")
```
